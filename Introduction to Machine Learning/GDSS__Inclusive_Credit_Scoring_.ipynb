**Our Goal**: Build a Machine Learning model that uses **alternative data** to predict creditworthiness, ensuring that the model is **inclusive** across different regions (e.g., Northern vs. Greater Accra) and **equitable** across genders.

## 🛠️ Phase 1: Environment Setup
We'll use standard Python libraries for the project.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("✅ Project environment ready!")
```

## 📊 Phase 2: Data Acquisition & EDA
We are using a synthetic dataset representing 2,000 Ghanaian small business owners.

```python
np.random.seed(2026)
size = 2000

data = {
    'Years_in_Business': np.random.randint(1, 15, size),
    'Monthly_MoMo_Volume_GHS': np.random.normal(5000, 2000, size).clip(500, 20000),
    'Utility_Bill_Consistency': np.random.uniform(0.5, 1.0, size), # % of bills paid on time
    'Region': np.random.choice(['Greater Accra', 'Ashanti', 'Northern', 'Western', 'Eastern', 'Volta'], size),
    'Gender': np.random.choice(['Female', 'Male'], size, p=[0.55, 0.45]), # Reflecting SME owner demographics
    'Sector': np.random.choice(['Agriculture', 'Retail', 'Tech', 'Craftsmanship'], size),
    'Previous_Default': np.random.choice([0, 1], size, p=[0.9, 0.1]),
    'Approved': np.random.choice([0, 1], size, p=[0.4, 0.6]) # Target Variable
}

df = pd.DataFrame(data)

# Quick look at the data distribution by Region
plt.figure(figsize=(10, 6))
sns.countplot(x='Region', hue='Approved', data=df, palette='Set2')
plt.title('Loan Approval Distribution by Region')
plt.show()
```

## ⚙️ Phase 3: Feature Engineering
As engineers, we know that raw data isn't always enough. Let's create a **'Trust Score'** based on MoMo volume and bill consistency.

```python
# Create a composite Trust Score
df['Trust_Score'] = (df['Monthly_MoMo_Volume_GHS'] / 20000) * 0.6 + (df['Utility_Bill_Consistency'] * 0.4)

# Convert Categorical to Numeric (Encoding)
df_ml = pd.get_dummies(df, columns=['Region', 'Gender', 'Sector'], drop_first=True)

print("New features created and categorical variables encoded.")
```

## 🤖 Phase 4: Model Development
We'll train a **Random Forest** to predict whether a loan should be `Approved`.

```python
X = df_ml.drop('Approved', axis=1)
y = df_ml['Approved']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print(f"Initial Model Accuracy: {accuracy_score(y_test, predictions)*100:.2f}%")
```

## ⚖️ Phase 5: The Ethical Audit (Crucial for GDSS 2026)
**Inclusivity Test**: Is our model biased against female entrepreneurs or specific regions? 

In a project-based intro, we don't just stop at accuracy. We look for **Fairness**.

```python
# Create a summary dataframe for the test set
audit_df = X_test.copy()
audit_df['Actual'] = y_test
audit_df['Predicted'] = predictions

# Check Approval Rate by Gender (Inferred from dummy columns if present)
if 'Gender_Male' in audit_df.columns:
    female_approval = audit_df[audit_df['Gender_Male'] == 0]['Predicted'].mean()
    male_approval = audit_df[audit_df['Gender_Male'] == 1]['Predicted'].mean()
    
    print(f"Approval Rate - Female: {female_approval*100:.1f}%")
    print(f"Approval Rate - Male: {male_approval*100:.1f}%")
    print(f"Disparity Gap: {abs(female_approval - male_approval)*100:.1f}%")

plt.figure(figsize=(8, 4))
plt.bar(['Female', 'Male'], [female_approval, male_approval], color=['pink', 'skyblue'])
plt.title('Gender-Based Inclusivity Check')
plt.ylabel('Approval Rate')
plt.show()
```
